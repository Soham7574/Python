### 09) WAP to count unique items in a list.

```python
l1 = list(map(int, input("Enter a List:").split()))
countUnique = 0
for i in range(len(l1)):
    flage = True
    for j in range(1,len(l1)):
        if l1[i] == l1[j]:
            if i != j:
                flage = False
                break;
    if flage:
        countUnique += 1
print(countUnique)
```

### 13) WAP to find second largest number in a list.

```python
l1 = [1,4,2,5,6,3]
max1 = l1[0]
secondMax = 0
for i in l1:
    if max1 < i:
        secondMax = max1
        max1 = i
    elif secondMax < i and i != max1:
        secondMax = i
print("Second Largest Element from list is:",secondMax)
```

### 14) WAP to extract elements with frequency greater than K.

```python
l1 = list(map(int, input("Enter a List:").split()))
k = int(input("k :"))
ans = []

for i in range(len(l1)):
    count = 1
    for j in range(0,len(l1)):
        if l1[i] == l1[j] and i != j:
            count += 1
    if count > k:
        ans.append(l1[i])
res = []
for i in ans:
    if i not in res:
        res.append(i)
print(res)
```
